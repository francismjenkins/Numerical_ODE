```python
# Solution to problem 25.5 Chapra & Canale, Numerical Methods for Engineers 6th Ed., McGraw Hill, New York, NY (2010)
# Author: Frank Jenkins

import math

def f(t, y):
    # solution for ODE dy/dt = y*sin^3(t)
    # return -1.947734*math.exp((1/12)*math.cos(3*t)-9*math.cos(t))
    return y*math.sin(t)**3

x = []
time = []

def heun(f, y0, a, b, h):  
    t, y, y1 = a, y0, 1
    while (t <= b):        
        #print("%6.3f %6.3f" % (t,y))
        x.append(y1)
        time.append(t)
        t += h
        y += h*f(t,y)
        y1 = y + (h/2)*(f(t,y) + y)
        #no corrector for this problem
        #y2 = (y+y1)/2

```

```python
c = heun(f, 1, 0, 3.0, 0.1)
print("Approximation at t = 3: ", x[-1])
import matplotlib.pyplot as plt

plt.plot(time, x)
```

```python
x1 = []
time1 = []

def rk2(f, y0, a, b, h):  
    t, y, y1 = a, y0, 0
    while (t <= b):        
        #print("%6.3f %6.3f" % (t,y))
        x1.append(y1)
        time1.append(t)
        t += h
        y += h*f(t,y)
        y1 = f(t + (3/4)*h, y0 + (3/4)*h*y)
```

```python
c = rk2(f, 1, 0, 3.0, 0.1)
print("Approximation at t = 3: ", x1[-1])
import matplotlib.pyplot as plt

plt.plot(time1, x1)
```
