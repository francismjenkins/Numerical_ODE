```python
# Solution to problem 25.5 Chapra & Canale, Numerical Methods for Engineers 6th Ed., McGraw Hill, New York, NY (2010)
# Author: Frank Jenkins

import math
import matplotlib.pyplot as plt
    
def f(t,y):
    yt = math.exp(-math.cos(t)+((math.cos(t)**3)/3)+(2/3))
    return yt

def dy(t,y):
    dy = y*math.sin(t)**3
    return dy

x = []
time = []

def heun(f, y0, a, b, h):  
    t, y = a, y0
    while (t <= b):        
        t += h
        y = f(t,y)
        y1 = f(t+h, h*f(t,y))
        x.append(y1)
        time.append(t)
    
```

```python
c = heun(f, 1, 0, 3.0, 0.1)
print("Approximation at t = 3: ", x[-1])
plt.plot(time, x)
plt.xlabel("Time")
plt.ylabel("Y")
```

```python
x2 = []
time2 = []

def rk4(y0, a, b, h):  
    t, y = a, y0
    while(t <= b):
        k1 = h*dy(t, y)
        k2 = h*dy(t + 0.5*h, y + 0.5*k1)
        k3 = h*dy(t + 0.5*h, y + 0.5*k2)
        k4 = h*dy(t + h, y + k3)
        y = y + (1/6)*(k1 + 2*k2 + 2*k3 + k4)
        t+=h
        x2.append(y)
        time2.append(t)
    return y
        
        
```
