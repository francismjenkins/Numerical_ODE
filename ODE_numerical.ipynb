```python
# Solution to problem 25.5 Chapra & Canale, Numerical Methods for Engineers 6th Ed., McGraw Hill, New York, NY (2010)
# Author: Frank Jenkins

import math
    
def f(t,y):
    y = math.exp(-math.cos(t)+((math.cos(t)**3)/3)+(2/3))
    return y

x = []
time = []

def heun(f, y0, a, b, h):  
    t, y = a, y0
    while (t <= b):        
        t += h
        y = f(t,y)
        y1 = f(t+h, h*f(t,y))
        x.append(y1)
        time.append(t)
    
```

```python
c = heun(f, 1, 0, 3.0, 0.1)
print("Approximation at t = 3: ", x[-1])
import matplotlib.pyplot as plt

plt.plot(time, x)
plt.xlabel("Time")
plt.ylabel("Y")
```

```python
c = rk2(f, 1, 0, 3.0, 0.1)
import matplotlib.pyplot as plt

plt.plot(time1, x1)
plt.xlabel("Time")
plt.ylabel("Y")
print("")
print("The approximation at t = 3: ", x1[-1])
```
